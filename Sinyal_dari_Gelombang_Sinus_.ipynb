```python
#O. Midiyanto(2001010021)
#pengaruh dari A,W,Fase
#Xt=A*sin(W*t+Fase)
#W=k*pi

import numpy as np
from matplotlib import pyplot as plt
from math import pi

t=np.linspace(-0.05,0.05,1000)

A=int(input("Masukkan Amplitudo = "))
W_input=int(input("Masukkan Frekuensi sudut (W) = "))
W=W_input*pi
Fase=int(input("Masukkan Fase = "))
Xt=A*np.sin(W*t+Fase)

plt.plot(t,Xt)
plt.xlabel("t / Waktu")
plt.ylabel("Amplitudo")
plt.xlim(-0.05,0.05)
plt.title("Plot dari Gelombang Sinus")
plt.show()
```
