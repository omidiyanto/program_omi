```python
import numpy as np
from matplotlib import pyplot as plt
from math import pi
```

```python
#Y=A*cos(w*t+fase)
#w=2*pi*f atau w=2*pi*n
#1. CT
A=5
f=10
t=np.linspace(-0.1,0.1,50)
fase=0
Y=A*np.cos(2*pi*f*t+fase)
plt.plot(t,Y)
plt.title("Sinyal CT")
plt.xlabel("Waktu(s)")
plt.ylabel("Amplitudo")
plt.show()
```

```python
#Y=A*cos(w*t+fase)
#w=2*pi*f atau w=2*pi*n
#2. DT
A=5
n=10
t=np.linspace(-0.1,0.1,50)
fase=0
Y=A*np.cos(2*pi*n*t+fase)
plt.stem(t,Y)
plt.title("Sinyal DT")
plt.xlabel("Waktu(s)")
plt.ylabel("Amplitudo")
plt.show()
```
