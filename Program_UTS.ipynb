```python
# program biseksi
def biseksi(x0, x1, e):
    iterasi = 1
    print('\n\n*** METODE BISEKSI ***')
    print('NO       x0              x1           f(x0)         f(x1)           xmid        f(xmid)')
    condition = True
    while condition:
        x2 = (x0 + x1)/2
        print('%d       %0.6f      %0.6f      %0.6f     %0.6f        %0.6f     %0.6f' % (iterasi, x0, x1, f(x0), f(x1), x2, f(x2)))

        if f(x0) * f(x2) < 0:
            x1 = x2
        else:
            x0 = x2

        iterasi = iterasi + 1
        condition = abs(f(x2)) > e
    print('\nAkar Persamaannya ialah : %0.8f' % x2)

# program regula falsi
def regulaFalsi(x0, x1, e):
    iterasi = 1
    print('\n\n*** METODE REGULA FALSI ***')
    print('NO       x0              x1           f(x0)         f(x1)           xmid        f(xmid)')
    condition = True
    while condition:
        x2 = x0 - (x1-x0) * f(x0)/(f(x1) - f(x0))
        print('%d       %0.6f      %0.6f      %0.6f     %0.6f        %0.6f     %0.6f' % (iterasi, x0, x1, f(x0), f(x1), x2, f(x2)))
        if f(x0) * f(x2) < 0:
            x1 = x2
        else:
            x0 = x2

        iterasi = iterasi + 1
        condition = abs(f(x2)) > e

    print('\nAkar Persamaannya ialah : %0.8f' % x2)

# program Newton Rhapson
def newtonRaphson(x0, e, N):
    print('\n\n*** METODE NEWTON RHAPSON ***')
    print('NO       x0              x1           f(x0)           f(x1)')
    step = 1
    flag = 1
    condition = True
    while condition:
        if g(x0) == 0.0:
            print('Dibagi 0 --> error!')
            break

        x1 = x0 - f(x0)/g(x0)
        print('%d       %0.6f       %0.6f       %0.6f       %0.6f' % (step, x0, x1, f(x0), f(x1)))
        x0 = x1
        step = step + 1

        if step > N:
            flag = 0
            break

        condition = abs(f(x1)) > e

    if flag == 1:
        print('\nAkar Persamaannya ialah : %0.8f' % x1)
    else:
        print('\nTidak Konvergen.')
```

```python
# No.1
def f(x):
    return x**3+5*x-1


def g(x):  # Turunan f(x) untuk digunakan di regula falsi
    return 3*x**2+5


x0 = 0  # xn
x1 = 0.5  # xn+1
N = 3  # untuk digunakan di Newton Rapshon
e = 0.005  # nilai toleransi/error
print('\n\t\t\t============  JAWABAN NO.1   ================')
if f(x0) * f(x1) > 0.0:
    print('Biseksi dan Regula Falsi gagal dieksekusi\nNilai xn dan xn+1 yang diberikan tidak mengurung akar persamaan\nCoba Lagi dengan nilai xn dan xn+1 yang berbeda.')
else:
    biseksi(x0, x1, e)
    regulaFalsi(x0, x1, e)

newtonRaphson(x0, e, N)
```
