```python
import sounddevice as sd
from scipy.io.wavfile import read, write
from IPython.display import Audio
from matplotlib import pyplot as plt
import time
```

```python
fs=44100
second=5
print("Recording start......")
record_voice=sd.rec(int(second*fs),samplerate=fs,channels=2)
for i in range (second,0,-1):
    print(i)
    time.sleep(1)
sd.wait()
print("Recording selesai, Tunggu sedang proses menampilkan sinyal")
write("AIUEO.wav",fs,record_voice)

fs_read,data=read("AIUEO.wav")
print("sampling rate: ",fs_read)
plt.figure()
plt.plot(data)
plt.title("AIUEO")
plt.xlabel("sample index")
plt.ylabel("wave form of test audio (Amplitudo)")
plt.show()
```
